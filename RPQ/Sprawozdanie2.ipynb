```python
import random
from Carlier import Carlier
from Task import Task
import matplotlib.pyplot as plt
from time import perf_counter
import Load
```

```python
tasks = Load.readData("Data/in100.txt")
cr_ = Carlier(10000)
print(f"Carlier  :{cr_.carlier(tasks)[0]}")
cr_ = Carlier()
print(f"Carlier z heapq :{cr_.carlier_heapq(tasks)[0]}")
```

```python
dataset = []
length = [50, 100, 200, 300, 500, 800, 1000, 1400, 1800]
for i in length:
    tmp = []
    for n in range(1,i+1):
        tmp.append(Task(id=n, r=random.randint(0, 1000), p=random.randint(0, 1000), q=random.randint(0, 1000)))
    dataset.append(tmp)

times1 = []
times2 = []
leng = [len(x) for x in dataset]

for data in dataset:
    start = perf_counter()
    cr = Carlier(100000)
    cr.carlier(data)
    times1.append(perf_counter()-start)

    start = perf_counter()
    cr = Carlier(100000)
    cr.carlier_heapq(data)
    times2.append(perf_counter()-start)

plt.plot(leng, times1, label="Bez kolejki")
plt.plot(leng, times2, label="Z kolejką")
plt.legend()
plt.grid()
plt.title("Czas wykonania")
plt.xlabel("Liczba zadań")
plt.ylabel("Czas")
plt.show()
```
